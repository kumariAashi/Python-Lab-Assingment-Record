```python
#find the mean, median, and mode of numbers entered by the user

# Take input from the user
data = input("Enter numbers separated by spaces: ")

# Convert input string into a list of numbers
numbers = [int(x) for x in data.split()]
print("Enter numbers separated by spaces: ",data)
```

```python
#Calculate mean, median, and mode 

# Mean
mean = sum(numbers) / len(numbers)

# Median
numbers.sort()
n = len(numbers)
if n % 2 == 0:
    median = (numbers[n//2 - 1] + numbers[n//2]) / 2
else:
    median = numbers[n//2]

# Mode
max_count = 0
mode = numbers[0]
for num in numbers:
    count = numbers.count(num)
    if count > max_count:
        max_count = count
        mode = num

# Display results
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```
