```python
#Given a list of items (possibly with duplicates)
#write a program that removes duplicates and displays the sorted list.

#Take input from user
items = input("Enter items separated by spaces: ").split()
print("Enter items separated by spaces: ",items)
```

```python
#Remove duplicates and sort

unique_sorted = sorted(set(items))
print("Unique & Sorted Items:", unique_sorted)
```
