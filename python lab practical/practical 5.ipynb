```python
#Write a Python program to reverse only the vowels in a given string
#keeping other characters in their original positions

#Take input from user
text = input("Enter sentence: ")
print("Enter sentence: ",text)
```

```python
#Reverse only vowels

vowels = "aeiouAEIOU"
v = [ch for ch in text if ch in vowels][::-1]
result = ""

for ch in text:
    if ch in vowels:
        result += v.pop(0)
    else:
        result += ch

print("Output:", result)
```
