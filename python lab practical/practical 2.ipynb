```python
#wap to count no of vowels in string

# Take input from the user
text = input("Enter a string: ")
print("Enter a  string:",text)
```

```python
# Initialize count
count = 0

# Define vowels
vowels = "aeiouAEIOU"

# Count vowels
for char in text:
    if char in vowels:
        count += 1

# Display result
print("Number of vowels:", count)
```
