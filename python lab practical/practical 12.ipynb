```python
#create a list of a student from which randomly select the CR & shuffle the list


import random

# Create a list of students
students = ["Aashi", "Aman", "Sagar", "Shreya" ]

# Randomly select the CR
cr = random.choice(students)

# Shuffle the list
random.shuffle(students)

# Output
print("Class Representative (CR):", cr)
print("Shuffled Student List:", students)
```
