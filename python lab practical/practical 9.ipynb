```python
#Read data from a CSV file containing employee details (name, department, salary) and display the average salary by department.
#Create CSV file

import csv

# Create and write employee data into CSV file
with open("employees.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Department", "Salary"])
    writer.writerow(["John", "IT", 50000])
    writer.writerow(["Mary", "IT", 55000])
    writer.writerow(["Alice", "HR", 48000])
    writer.writerow(["Bob", "HR", 52000])

print("employees.csv file created successfully!")
```

```python
#Read CSV and calculate average salary by department

import csv

# Dictionaries to store total salary and count
dept_salary = {}
dept_count = {}

# Read CSV file
with open("employees.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        dept = row["Department"]
        salary = int(row["Salary"])
        dept_salary[dept] = dept_salary.get(dept, 0) + salary
        dept_count[dept] = dept_count.get(dept, 0) + 1

# Display average salary by department
print("Average Salary by Department:")
for dept in dept_salary:
    avg = dept_salary[dept] / dept_count[dept]
    print(f"{dept}: {avg}")
```
