```python
#wap to reverse a no 123456 to 654321

# Take input from the user
num = int(input("Enter a number: "))
print("Enter a number:" , num)
```

```python
#Reverse the number

# Initialize reverse number
rev = 0

# Logic to reverse digits
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

# Display the reversed number
print("Reversed number:", rev)
```
