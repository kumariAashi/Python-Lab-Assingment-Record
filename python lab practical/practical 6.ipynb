```python
#Create a script that takes an integer and
# displays its binary, octal, and hexadecimal representations neatly formatted

#Take input from user
num = int(input("Enter a number: "))
print("Enter a number: ",num)

```

```python
#display conversion

print("Binary:", bin(num)[2:])
print("Octal:", oct(num)[2:])
print("Hexadecimal:", hex(num)[2:].upper())
```
